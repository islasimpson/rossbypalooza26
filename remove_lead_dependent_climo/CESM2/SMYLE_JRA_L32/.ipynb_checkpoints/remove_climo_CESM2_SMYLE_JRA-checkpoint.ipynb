```python
import xarray as xr
import numpy as np
```

### Information on data location and variable name

```python
basepath="/glade/campaign/cgd/cas/islas/python_savs/rossbypalooza26/processing/CESM2/SMYLE_JRA_L32/"
var="slp"

outpath="/glade/campaign/cgd/cas/islas/python_savs/rossbypalooza26/remove_lead_dependent_climo/CESM2/SMYLE_JRA_L32/"
```

### Years used to define the climatology

```python
yclimo1 = 1993
yclimo2 = 2016
```

### Open data remove the lead dependent climatology

```python
dat = xr.open_dataset(basepath+"CESM2_SMYLE_JRA_L32"+var+"_1993_2022.nc")
```

```python
### Method 1
# Remove the climatology for all members
dat_anoms = dat - dat.mean(['year','member'])
dat_anoms = dat_anoms.transpose('member','year','lead','lat','lon')
dat_anoms.to_netcdf(outpath+var+'_anoms_ALLM_1993_2022.nc')
del(dat_anoms)
```

```python
### Method 2
# Remove the climatology for each member individually
dat_anoms = dat - dat.mean(['year'])
dat_anoms = dat_anoms.transpose('member','year','lead','lat','lon')
dat_anoms.to_netcdf(outpath+var+'_anoms_INDM_1993_2022.nc')
```
