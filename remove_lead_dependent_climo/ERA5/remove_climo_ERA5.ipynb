```python
import xarray as xr
import numpy as np
```

### Information on data location and variable name

```python
basepath="/glade/campaign/cgd/cas/islas/python_savs/rossbypalooza26/processing/ERA5/"
var="slp"

outpath="/glade/campaign/cgd/cas/islas/python_savs/rossbypalooza26/remove_lead_dependent_climo/ERA5/"
```

### Years used to define the climatology

```python
yclimo1 = 1993
yclimo2 = 2016
```

### Open data remove the lead dependent climatology

```python
dat = xr.open_dataset(basepath+"ERA5_monthly_"+var+"_1993_2025.nc")

# remove the lead dependent climatology
dat_anoms = dat - dat.sel(year=slice(yclimo1,yclimo2)).mean('year')

dat_anoms.to_netcdf(outpath+var+'_anoms_1993_2025.nc')
```
